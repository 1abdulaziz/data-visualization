```python
#magic command to display matplotlib plots inline within the ipython notebook webpage
%matplotlib inline

#import necessary modules
import pandas as pd, numpy as np, matplotlib.pyplot as plt, matplotlib as mpl
import math

# configure the pandas matplotlib style to make the default plots look a bit nicer
pd.options.display.mpl_style = 'default'
```

# PART 1: PREP THE DATA

```python
st = pd.read_csv('data/summer-travel-gps-full.csv', index_col='date', parse_dates=True)
st.tail()
```

```python
tc = pd.read_csv('data/traffic-volume.csv', index_col='date', parse_dates=True)
tc.dropna(inplace=True)
tc.tail()
```

```python
# weekday determines which day of the week a date is. 0 is sunday and 6 is saturday. M-F are 1-5.
# use bitwise OR to create a boolean array of which rows are a weekend date
weekend_mask = (tc.index.weekday==6) | (tc.index.weekday==0)
weekends = tc[weekend_mask]
weekdays = tc[~ weekend_mask]
```

```python
hourly_share = pd.DataFrame(columns=['weekday', 'weekend'])

# calculate what share of weekday traffic volume each hour has
weekday_hourly_avg = weekdays.groupby(weekdays.index.hour).mean()
hourly_share['weekday'] = weekday_hourly_avg / sum(weekday_hourly_avg['volume'])

# calculate what share of weekend traffic volume each hour has
weekend_hourly_avg = weekends.groupby(weekends.index.hour).mean()
hourly_share['weekend'] = weekend_hourly_avg / sum(weekend_hourly_avg['volume'])

hourly_share.index = [s + ':00' for s in hourly_share.index.astype(str)]

# calculate what share of total traffic volume each day of the week has
daily_avg = tc.groupby(tc.index.weekday).mean() * 24
daily_share = daily_avg / sum(daily_avg['volume'])
daily_share.index = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday']
```

# PART 4: PIE CHARTS

```python
# ugly default matplotlib style pie chart
daily_share.plot(kind='pie', y='volume', legend=False)
```

```python
# function to produce more beautiful pie charts with matplotlib
def gbplot_pie(fractions, #values for the wedges
              labels, #labels for the wedges
              title = '', #title of the pie chart
              cm_name = 'Pastel1', #name of the matplotlib colormap to use
              autopct = '%1.1f%%', #format the value text on each pie wedge
              labeldistance = 1.05, #where to place wedge labels in relation to pie wedges
              shadow = True, #shadow around the pie
              startangle = 90, #rotate 90 degrees to start the top of the data set on the top of the pie
              edgecolor = 'w', #color of pie wedge edges
              width = 8, #width of the figure in inches
              height = 8, #height of the figure in inches
              grouping_threshold = None, #group all wedges below this value into one wedge
              grouping_label = None): #what the label the grouped wedge
    
    # if the user passed a threshold value, group all fractions lower than it into one 'misc' pie wedge
    if not grouping_threshold==None:
        
        # if user didn't pass a label, apply a default text
        if grouping_label == None:
            grouping_label = 'Others'

        # select the rows greater than the cutoff value
        row_mask = fractions > grouping_threshold
        meets_threshold = fractions[row_mask]

        # group all other rows below the cutoff value
        all_others = pd.Series(fractions[~row_mask].sum())
        all_others.index = [grouping_label]

        # append the grouped row to the bottom of the rows to display
        fractions = meets_threshold.append(all_others)
        labels = fractions.index
    
    # get the color map then pull 1 color from it for each pie wedge we'll draw
    color_map = mpl.cm.get_cmap(cm_name)
    num_of_colors = len(fractions)
    colors = color_map([x/float(num_of_colors) for x in range(num_of_colors)])
    
    # create the figure and an axis to plot on
    fig, ax = plt.subplots(figsize=[width, height])
    
    # plot the pie
    wedges = ax.pie(fractions, 
                    labels = labels, 
                    labeldistance = labeldistance,
                    autopct = autopct,
                    colors = colors,
                    shadow = shadow, 
                    startangle = startangle)
    
    # change the edgecolor for each wedge
    for wedge in wedges[0]:
        wedge.set_edgecolor(edgecolor)
    
    # set the title and show the plot
    ax.set_title(title)
    plt.show()
```

```python
# nicely styled pie charts
gbplot_pie(fractions = daily_share['volume'],
          labels = daily_share.index,
          title = 'Traffic volume share, by day of week')
```

```python
# ugly default matplotlib style pie chart
hourly_share.plot(kind='pie', y='weekday', legend=False)
```

```python
# use my grouping_threshold argument to group all values below some threshold to make chart much less messy
gbplot_pie(fractions = hourly_share['weekday'],
          labels = hourly_share.index,
          title = 'Weekday traffic volume share, by hour',
          grouping_threshold = 0.04,
          grouping_label = 'All Other Times')
```

```python
# show absolute count values instead of relative percentages
gbplot_pie(fractions = hourly_share['weekend'],
          labels = hourly_share.index,
          autopct=lambda(p): '{:.0f}'.format(p * len(hourly_share['weekend'])),
          title = 'Weekend traffic volume, by hour',
          grouping_threshold = 0.03)
```
